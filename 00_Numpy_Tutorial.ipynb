# NumPy Tutorial for Machine Learning

This tutorial introduces NumPy, the fundamental library for numerical computing in Python. NumPy is essential for Machine Learning because it provides fast array operations that are the building blocks for neural networks.

---

## Why NumPy?

| Python Lists | NumPy Arrays |
|--------------|-------------|
| Slow (loops) | Fast (vectorized) |
| General purpose | Optimized for numbers |
| No math operations | Built-in math operations |

**Example:** Adding two lists of 1 million numbers:
- Python lists: ~100ms
- NumPy arrays: ~1ms (100x faster!)

The speed of NumPy stems from its use of contiguous memory layout with fixed data types, enabling vectorized operations executed in **pre-compiled C code** that bypass Python's interpreter overhead and per-element type checking.

# Part 6: Matrix Operations (Essential for Neural Networks!)

Matrix multiplication is THE core operation in neural networks.

## 6.1 Dot Product

````python
# Dot product of two vectors
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# a · b = 1*4 + 2*5 + 3*6 = 4 + 10 + 18 = 32
dot_product = np.dot(a, b)

print(f"a = {a}")
print(f"b = {b}")
print(f"a · b = {dot_product}")
````

## 6.2 Matrix Multiplication

**Rule:** For `A @ B`, the number of columns in A must equal the number of rows in B.

If A is `(m, n)` and B is `(n, p)`, then A @ B is `(m, p)`.

````python
# Matrix-vector multiplication (used in neural networks!)
W = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])  # Shape: (3, 2)

x = np.array([1, 2])  # Shape: (2,)

# W @ x: (3, 2) @ (2,) = (3,)
result = np.dot(W, x)

print(f"W (3x2):\n{W}")
print(f"\nx (2,): {x}")
print(f"\nW @ x = {result}")
print(f"Result shape: {result.shape}")
````

````python
# Matrix-matrix multiplication
A = np.array([[1, 2], [3, 4]])  # (2, 2)
B = np.array([[5, 6], [7, 8]])  # (2, 2)

# You can use np.dot() or the @ operator
C = A @ B  # Same as np.dot(A, B)

print(f"A:\n{A}")
print(f"\nB:\n{B}")
print(f"\nA @ B:\n{C}")
````

# Part 7: Reshaping Arrays

Often you need to change the shape of arrays without changing the data.

````python
# Reshape a 1D array to 2D
arr = np.arange(12)  # [0, 1, 2, ..., 11]
print(f"Original: {arr}")
print(f"Shape: {arr.shape}")

# Reshape to 3x4
reshaped = arr.reshape(3, 4)
print(f"\nReshaped (3x4):\n{reshaped}")

# Reshape to 4x3
reshaped2 = arr.reshape(4, 3)
print(f"\nReshaped (4x3):\n{reshaped2}")
````

````python
# Using -1 to auto-calculate one dimension
arr = np.arange(12)

# "Give me 3 rows, calculate columns automatically"
reshaped = arr.reshape(3, -1)
print(f"reshape(3, -1):\n{reshaped}")

# "Give me 4 columns, calculate rows automatically"
reshaped2 = arr.reshape(-1, 4)
print(f"\nreshape(-1, 4):\n{reshaped2}")
````

````python
# Flatten: Convert any array to 1D
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(f"Matrix:\n{matrix}")
print(f"\nFlattened: {matrix.flatten()}")

# This is how images are fed into neural networks!
# A 28x28 image becomes a 784-element vector
````

# Part 9: Boolean Indexing

Select elements based on conditions.

````python
arr = np.array([1, -2, 3, -4, 5, -6])

print(f"Array: {arr}")
print(f"\nPositive mask: {arr > 0}")
print(f"Positive elements: {arr[arr > 0]}")
print(f"Negative elements: {arr[arr < 0]}")
````

````python
# np.where: conditional selection
arr = np.array([-2, -1, 0, 1, 2])

# ReLU using np.where!
relu_result = np.where(arr > 0, arr, 0)
print(f"Array: {arr}")
print(f"ReLU: {relu_result}")

# Or simply use np.maximum
print(f"ReLU (using maximum): {np.maximum(0, arr)}")
````

# Summary: NumPy Operations for ML

| Operation | NumPy Function | ML Use Case |
|-----------|---------------|-------------|
| Create array | `np.array()`, `np.zeros()`, `np.ones()` | Initialize data |
| Random array | `np.random.randn()` | Initialize weights |
| Matrix multiply | `np.dot()` or `@` | Layer computation |
| Element-wise | `+`, `-`, `*`, `/` | Bias addition, scaling |
| Maximum | `np.maximum(0, x)` | ReLU activation |
| Exponential | `np.exp()` | Softmax activation |
| Sum | `np.sum()` | Softmax normalization |
| Argmax | `np.argmax()` | Get predicted class |
| Reshape | `.reshape()` | Flatten images |
